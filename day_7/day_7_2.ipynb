```python
#普通散点图
import visdom
import numpy as np

vis = visdom.Visdom()#这会启动一个与Visdom服务器的连接

Y = np.random.rand(100)
old_scatter = vis.scatter(
    X=np.random.rand(100, 2),
    Y=(Y[Y > 0] + 1.5).astype(int),
    opts=dict(
        legend=['Didnt', 'Update'],
        xtickmin=-50,
        xtickmax=50,
        xtickstep=0.5,
        ytickmin=-50,
        ytickmax=50,
        ytickstep=0.5,
        markersymbol='cross-thin-open',
    ),
)

vis.update_window_opts(
    win=old_scatter,
    opts=dict(
        legend=['2019年', '2020年'],
        xtickmin=0,
        xtickmax=1,
        xtickstep=0.5,
        ytickmin=0,
        ytickmax=1,
        ytickstep=0.5,
        markersymbol='cross-thin-open',
    ),
)
```

```python
#带文本标签的散点图
import visdom
import numpy as np

vis.scatter(
    X=np.random.rand(6, 2),
    opts=dict(
        textlabels=['Label %d' % (i + 1) for i in range(6)]
    )
)
```

```python
#三维散点图
import visdom
import numpy as np
Y = np.random.rand(100)
vis.scatter(
    X=np.random.rand(100, 3),
    Y=(Y + 1.5).astype(int),
    opts=dict(
        legend=['男性', '女性'],
        markersize=5,#设置每个散点的大小为5
        xtickmin=0,
        xtickmax=2,
        xlabel='数量',
        xtickvals=[0, 0.75, 1.6, 2],
        ytickmin=0,
        ytickmax=2,
        ytickstep=0.5,
        ztickmin=0,
        ztickmax=1,
        ztickstep=0.5,
    )
)
```

```python
#线型图
import visdom
import numpy as np

Y = np.linspace(-5, 5, 100)#-5到5的100个点
vis.line(
    Y=np.column_stack((Y * Y, np.sqrt(Y + 5))),#创建两个函数
    X=np.column_stack((Y, Y)),
    opts=dict(markers=False),
)
```

```python
#实线、虚线等不同线
import visdom
import numpy as np

win = vis.line(
    X=np.column_stack((
        np.arange(0, 10),
        np.arange(0, 10),
        np.arange(0, 10),
    )),
    Y=np.column_stack((
        np.linspace(5, 10, 10),
        np.linspace(5, 10, 10) + 5,
        np.linspace(5, 10, 10) + 10,
    )),
    opts={
        'dash': np.array(['solid', 'dash', 'dashdot']),
        'linecolor': np.array([
            [255, 191, 255],
            [0, 191, 255],
            [255, 0, 0],
        ]),
        'title': '不同类型的线'
    }
)

vis.line(
    X=np.arange(0, 10),
    Y=np.linspace(5, 10, 10) + 15,
    win=win,
    name='4',
    update='insert',
    opts={
        'linecolor': np.array([
            [255, 0, 0],
        ]),
        'dash': np.array(['dot']),
    }
)
```

```python
#堆叠区域
import visdom
import numpy as np

Y = np.linspace(0, 4, 200)
win = vis.line(
    Y=np.column_stack((np.sqrt(Y), np.sqrt(Y) + 2)),
    X=np.column_stack((Y, Y)),
    opts=dict(
        fillarea=True, # 启用填充区域
        showlegend=False,
        width=380, # 图表宽度（像素）
        height=330,  # 图表高度（像素）
        ytype='log', # Y轴使用对数刻度
        title='堆积面积图',
        marginleft=30,# 左边距
        marginright=30,
        marginbottom=80,
        margintop=30,
    ),
)
```

```python
#茎叶图
import math
import visdom
import numpy as np

Y = np.linspace(0, 2 * math.pi, 70)
X = np.column_stack((np.sin(Y), np.cos(Y)))
vis.stem(
    X=X,
    Y=Y,
    opts=dict(legend=['正弦函数', '余弦函数'])
)
```

```python
#热力图
import visdom
import numpy as np

vis.heatmap(
    X=np.outer(np.arange(1, 6), np.arange(1, 11)),
    opts=dict(
        columnnames=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'],
        rownames=['y1', 'y2', 'y3', 'y4', 'y5'],
        colormap='Hot',
    )
)
```

```python
#条形图
import visdom
import numpy as np

vis.bar(
    X=np.abs(np.random.rand(4, 3)),
    opts=dict(
        stacked=True,#启用堆叠模式
        legend=['低价值客户', '一般价值客户', '高价值客户'],
        rownames=['2017', '2018', '2019', '2020']

    )
)
```

```python
#箱形图
import visdom
import numpy as np

X = np.random.rand(100, 2)
X[:, 1] += 2
vis.boxplot(
    X=X,
    opts=dict(legend=['男性', '女性'])
)
```

```python
#曲面图
import visdom
import numpy as np

X = np.random.rand(50, 2)
X[:, 1] += 1
vis.surf(X=X, opts=dict(colormap='Viridis'))
```

```python
#等高线
import visdom
import numpy as np

x = np.tile(np.arange(1, 81), (80, 1))
y = x.transpose()#将x矩阵转置
X = np.exp((((x - 40) ** 1) + ((y - 40) ** 1)) / -(20.0 ** 2))
vis.contour(X=X, opts=dict(colormap='Viridis'))
```
