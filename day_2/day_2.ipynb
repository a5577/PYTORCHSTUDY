```python
#数组直接创造张量
import numpy as np
import torch
#arr = np.zeros((3,3))
arr = np.ones((3,3))
print("arr的数据类型是：",arr.dtype)
t1 = torch.tensor(arr)
t2 = torch.tensor(arr,device = "cpu")
print(t1)
print(t2)
```

```python
#从numpy创作张量
arr = np.array([[1,2,3],[2,3,4]])
t1 = torch.from_numpy(arr)# torch.from_numpy与array共存同时会发生变化
print("原始数组和张量")
print(arr)
print(t1)
#-------------------------------------------------------------------
print("修改后的数组和张量")
t1[1,2] = 5
print(arr)
print(t1)
```

```python
#---------------------概率分布创建张量-----------------------
mean = torch.arange(1,5,dtype=torch.float)
std = torch.arange(1,5,dtype=torch.float)#均为张量
t = torch.normal(mean,std)
print(mean)
print(std)
print(t)#正态分布随机生成数字
```

```python
#------------------激活函数-------------------------
import numpy as np
import matplotlib.pyplot as plt #画图
def sigmoid(x):
    return 1./(1.+np.exp(-x))
def plot_sigmoid():
    x = np.arange(-10,10,0.1)
    y = sigmoid(x)
    plt.plot(x,y)
    plt.show()
if __name__ == '__main__':#如果这个Python文件被直接运行（而不是被导入），那么就执行 plot_sigmoid() 函数。
    plot_sigmoid()
```

```python
import numpy as np
import matplotlib.pyplot as plt #画图
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def plot_tanh():
    x = np.arange(-10,10,0.1)
    y = tanh(x)
    plt.plot(x,y)
    plt.show()
if __name__ == '__main__':#如果这个Python文件被直接运行（而不是被导入），那么就执行 plot_sigmoid() 函数。
    plot_tanh()
```

```python
import numpy as np
import matplotlib.pyplot as plt #画图
def ReLU(x):
    return np.maximum(0,x)
def plot_ReLU():
    x = np.arange(-10,10,0.1)
    y = ReLU(x)
    plt.plot(x,y)
    plt.show()
if __name__ == '__main__':#如果这个Python文件被直接运行（而不是被导入），那么就执行 plot_sigmoid() 函数。
    plot_ReLU()
```

```python
#--------------损失函数-----------------
#L1、
loss = torch.nn.L1Loss(reduction = 'mean')
input = torch.tensor([1.0,3.0,5.0,7.0])
target = torch.tensor([2.0,4.0,6.0,8.0])
output = loss(input,target)
print(output)
```

```python
#均方误差损失函数
loss = torch.nn.MSELoss(reduction = 'mean')
input = torch.tensor([1.0,3.0,5.0,7.0])
target = torch.tensor([3.0,5.0,7.0,9.0])
output = loss(input,target)
print(output)
```

```python
#交叉熵损失函数
entroy = torch.nn.CrossEntropyLoss()
input = torch.tensor([[0.2546,-0.2682,0.1258,-0.1012]])
target = torch.tensor([0])
output = entroy(input,target)
print(output)
```

```python
#余弦相似度损失
import torch
a = torch.tensor([1.0,2.0,3.0,4.0])
b = torch.tensor([4.1,6.1,7.1,8.1])
similarity = torch.cosine_similarity(a,b,dim = 0)
loss = 1-similarity
print(loss)
```
